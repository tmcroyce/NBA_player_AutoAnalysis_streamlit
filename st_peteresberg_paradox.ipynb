```python
import random

def st_petersburg_paradox():
    winnings = 2
    flips = 0
    while random.random() < 0.5:
        winnings *= 2
        flips += 1
    return winnings, flips


def run_sim_100():
    winnings_distribution = []
    num_simulations = 100000
    total_winnings = 0
    for i in range(num_simulations):
        winnings, flips = st_petersburg_paradox()
        total_winnings += winnings
        winnings_distribution.append(winnings)

    return(total_winnings/num_simulations)


```

```python
run_sim_100()
```

```python
# run run_sim_100() 10000 times and plot the distribution of winnings
import matplotlib.pyplot as plt
import numpy as np

average_winnings = []
for i in range(1000):
    average_winnings.append(run_sim_100())

# plot the distribution of winnings wwith plotly
import plotly.graph_objects as go

fig = go.Figure(data=[go.Histogram(x=average_winnings)])
fig.show()

# add cdf
import plotly.express as px

fig = px.histogram(average_winnings, nbins=100, histnorm='probability density', cumulative=True, title='CDF of average winnings')
fig.show()



```

```python
# say I have 10,000 dollars and I want to play the st petersburg paradox at 10 dollars a flip.
# simulate this
import random

def st_petersburg_paradox():
    winnings = 2
    flips = 0
    while random.random() < 0.5:
        winnings *= 2
        flips += 1
    return winnings, flips

starting_money = 10000
bet = 20
num_simulations = 100000
total_winnings = 0
for i in range(num_simulations):
    winnings, flips = st_petersburg_paradox()
    total_winnings += winnings
    starting_money -= bet
    starting_money += winnings
    # if starting_money < 0: game ends
    if starting_money < 0:
        print('YOU WENT BROKE FIRST')
        # stop the simulation
        break

print(starting_money)

```

```python
def run_it(bet,simulations):
    starting_money = 10000
    num_simulations = simulations
    total_winnings = 0
    for i in range(num_simulations):
        winnings, flips = st_petersburg_paradox()
        total_winnings += winnings
        starting_money -= bet
        starting_money += winnings
        # if starting_money < 0: game ends

    return(starting_money)
```

```python
def run_it_abunch(bet, simulations, sim2):
    results = []
    for i in range(sim2):
        results.append(run_it(bet,simulations))
    return results
```

```python
ten_dollas = run_it_abunch(10,100000,1000)
# number of negative results
len([i for i in ten_dollas if i < 0])
```
